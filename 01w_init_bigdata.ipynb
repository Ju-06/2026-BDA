```python
import pandas as pd
pop = pd.read_csv('popmonth.csv', encoding='cp949', index_col=0, thousands=',')
pop
```

```python
pop.columns = ['총인구수', '세대수', '세대당인구', '남자인구수', '여자인구수', '남여비율']
pop.head()
```

```python
pop.index[1]
```

```python
pop.index = [pop.index[i].split()[0] for i in range(len(pop))]
pop

# 0
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(pop['총인구수'])
plt.xticks(rotation=90) # x축 글씨를 90도 회전
plt.show()
```

```python
pop.sort_values('총인구수') #작은 것 부터 정렬
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pop1 = pop.sort_values('총인구수')
sns.barplot(x=pop1.index, y='총인구수', data=pop1)
plt.xticks(rotation=90)
plt.show()
```
